# **1. Import Modules**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# **2. Import Data**

```python
filePath = "Data/Super-Store-2021.xlsx"
```

```python
# df_raw = pd.read_excel(filePath, sheet_name="Sample - Superstore_Orders")
df_raw = pd.read_excel(filePath)
df_raw.info()
```

# **3. Create KPIs**

```python
#Total Revenue
df = df_raw.copy()

df['Month'] = df['Order Date'].dt.year.astype(str) + ' - ' + df['Order Date'].dt.month.astype(str)
df['Month']
```

## **3.1. Total Revenue**

```python
df_revenue = df.groupby('Month')['Sales'].sum().reset_index()
df_revenue
```

```python
plt.figure(figsize=(30, 10))
plt.plot(df_revenue['Month'], df_revenue['Sales'], c='red', marker='o', linestyle='solid', linewidth=2, markersize=12)
plt.title('Total Revenue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## **3.2. Total Profit**

```python
df_profit = df.groupby('Month')['Profit'].sum().reset_index()
df_profit
```

```python
plt.figure(figsize=(30, 10))
plt.plot(df_profit['Month'], df_profit['Profit'], c='red', marker='o', linestyle='solid', linewidth=2, markersize=12)
plt.title('Total Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

## **3.3. Profit Ratio**

```python
df_profit_ratio = df_revenue.merge(df_profit, on='Month')
df_profit_ratio['Profit Ratio'] = df_profit_ratio['Profit'] / df_profit_ratio['Sales']
df_profit_ratio
```

```python
plt.figure(figsize=(30, 10))
plt.plot(df_profit_ratio['Month'], df_profit_ratio['Profit Ratio'], c='red', marker='o', linestyle='solid', linewidth=2, markersize=12)
plt.title('Profit Ratio')
plt.xlabel('Month')
plt.ylabel('Profit Ratio')
plt.show()
```

## **3.4. Customer**

```python
#count distinct customer
df_customer = df.groupby('Customer ID')['Customer Name'].nunique().reset_index()
df_customer
```

```python
df.groupby('Customer ID')['Customer Name'].nunique().value_counts()
```

## **3.5 Sales per Customer**

```python
df_sales_per_customer = df.groupby('Customer ID')['Sales'].sum().reset_index()
df_sales_per_customer
```

```python
df.groupby('Customer ID')['Quantity'].sum().reset_index()
```

# **4. Visualize**

```python
#bar chart visualize Total Revenue
plt.figure(figsize=(30, 10))
sns.barplot(x='Customer ID', y='Sales', data=df_sales_per_customer)
```
